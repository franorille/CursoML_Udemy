```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://es.m.wikipedia.org/wiki/Oviedo"
```

```python
html = requests.get(url).text
html
```

```python
soup = BeautifulSoup(html,"lxml")
soup
```

```python
print(soup.prettify())
```

```python
gobiernos_oviedo = soup.find('table',attrs={'class' : 'center wikitable col1cen'})
gobiernos_oviedo
```

```python
tbody = gobiernos_oviedo.tbody
tbody

```

```python
columnas = []
for th in tbody.find_all("th"):
    columnas.append(th.text.replace("\n",""))

columnas
```

```python
datos = []
for tr in tbody.find_all("tr"):
    row = []
    for td in tr.find_all("td"):
        row.append(td.text.replace("\n",""))
    datos.append(row)
datos
```

```python
df = pd.DataFrame(datos, columns=columnas)
df
```
