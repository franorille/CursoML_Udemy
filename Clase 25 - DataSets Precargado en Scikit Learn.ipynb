Luego tienen otros datasets más grandes también para practicar, más complejos aunque también preparados para poder cargarlos mediante una línea de código. 

```python
# lo primero que hay que hacer es importar la librería sklearn
from sklearn.datasets import load_breast_cancer # de esta forma ya tenemos el método preparado para lanzarlo
```

```python
""" load_breast_cancer(tiene 2 posibles parámetros) 1º return:_X_u=False por defecto. si es true devuelve los datos mediante un formato (data, target) as_frame=False  devuelve los datos en formato DataFrame
# Aunque realmente no devuelve exactamente un DF de pandas sino que es un diccionario con más cosas dentro con varias claves. 
Por lo que debe sacarse lo que nos interesa, que es el atributo frame, que es el que viene en formato dataframe
"""

data = load_breast_cancer(as_frame=True)
```

```python
data
```

```python
df = data.frame #En variable df cargamos el dataFrame en formato de pandas,
# ya que la variable data es un diccionario que incluye más claves a parte del dataframe. 
df
```

```python
#la variable target toma valor 0 si es cancer maligno o 1 si es benigno el cancer
sum(df.target == 0) # hay 212 que son malignos
```
