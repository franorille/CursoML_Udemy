```python
""" método de instancia que permite ver las primeras filas del df. Por defecto muestra 5 primeras filas, pero se puede cambiar, 
ya sea poniendo un número concreto n=nº o un porcentaje sobre el total del df 
"""
df.head()
```

```python
df.tail(n=2)
```

```python
"""
Extraer muestra de filas aleatorias. df.sample()
Es al azar, cada vez que recargas la celda será una diferente, aunque se puedon modificar ciertos parámetros.
n=nº filas a mostrar
frac=0. porcentaje del total a mostrar. Este parámetro es incompatible con n
random_tate=nº fija un punto de partida a partir del cual siempre obtendremos los mismos resultados. Se trata de fijar una semilla. por lo que siempre obtendremos los mismos datos
                la seleccion ha sido aleatorio, pero permite replicar resultados.
                Este parámetro está disponible en otros muchos métodos.
"""
df.sample()
```

```python
df.sample(4)
```

```python
df.sample(frac=0.01)
```

```python
df.sample(n=5, random_state=10)
```
