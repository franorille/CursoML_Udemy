<h4> Selección de columnas </h4>

<h3> SELECCION BASADA EN ETIQUETAS CON .loc </h3>

Método basado en etiquetas (nombres de columnas, o índices generados automáticamente o generados por nosotros), no en posiciones. 
Es util porque no en todos los casos se van a conocer su posicion exacta o el índice concreto de un valor, pero igual si que conocemos sus etiquetas, como el eje de las columnas o filas.
Para ello usaremos el método .loc

```python
df = pd.DataFrame({"Columna1":10,"Columna2":20,"Columna3":30},index=["Fila1","Fila2","Fila3",]) # Creación de un df con su constructor, pasándole un diccionario con las columnas y las filas
df
# generamos un índice manual dando nosotros las etiquetas de las filas y columnas
```

```python
# En este enfoque hacemos referencia a las etiquetas de las filas y columnas
df.loc['Fila1',"Columna1"]
```

```python
# Seleccion de todas las filas de las columnas education, martial-status y relationships.
# con el operador : se obtienen todas las filas, y como segundo parámetro se pasa una lista con los nombres de las columnas
adult_data.loc[:,['education','marital-status','relationship']]
```

```python
# Primer valor de la columma education
adult_data.loc[0,'education'] # el cero sin comillas, es el índice de la primera fila y luego el nombre de la columna.
```

Ojo, Metodo iloc incluye el primer valor de los corchetes pero excluye el último elemento-[0:10] incluye valores desde 0 hasta 9
Método iloc no admite etiquetas, osea Strings

Método loc incluye ambos valores inicial y final

```python
adult_data.loc[0:10, ['education','marital-status','relationship']] # devuelve desde fila 0 hasta 10 incluida -- 11 filas
```
